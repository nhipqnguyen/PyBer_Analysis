```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Set the x limit between Jan and June.
plt.xlim("Jan", "June")
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
# Change the figure size to 20 pixels wide by 10 pixels high.
plt.figure(figsize=(20, 10))
```

```python
# Using the object-oriented approach, make the following changes to your line chart:
    ## Change the line color to green.
    ## Change the marker to a diamond.
    ## Change the line width to 2 points.
    ## Add a legend for the city of Boston.
    ## Add a title and axes labels.
    ## Add a y-axis limit.
    ## Add grid lines.
    
# Create the plot.
ax = plt.axes()
ax.plot(x_axis, y_axis, marker = 'd', color = "green", linewidth = "2", label = 'Boston')
# Add the legend.
ax.legend()
# Create labels for the x and y axes.
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
# Create a title.
ax.set_title("PyBer Fare by Month")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Add a grid.
ax.grid()
```

```python
# maek the bubbles bigger
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using the object-oriented approach, make the following changes to your bubble chart:
    # Change the color of the markers to sky blue.
    # Change the size of the markers to 5 times each data point.
    # Make the color 20% transparent.
    # Add a black edge color to the circles.
    # Make the linewidth of the circles 2 points.
    # Add a legend for the city of Boston.
    # Switch the axis so that the Fare($) data is on the x-axis.
ax = plt.axes()
ax.scatter(y_axis, x_axis, color = "skyblue", linewidths = 2, edgecolors = "black", s = [i * 5 for i in y_axis], label = "Boston",alpha = 0.2)
    # Add a title.
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
    # Add a limit for the x-axis data.
ax.set_xlim(0,50)
    # Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# increase size of pie chart
plt.subplots(figsize=(8, 8))
# pop out July, add percentage of each wedge (1 decimal place)
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```
